### Problem 2 : Uni-Variate Linear Regression

**a) Partition the dataset randomly into Train and Test datasets  70% for Train and 30% for Test. On the Train set, fit three uni-variate Linear Regression models for Length against each of the three other variables. For each such linear model, print the coefficients of the model you fit, and plot the regression line on a scatterplot.**

````python
## Using Random Splitting
from sklearn.model_selection import train_test_split
quiz_data_train, quiz_data_test = train_test_split(quiz_data, test_size = 0.3)
````

````python
quiz_data_train_y = pd.DataFrame(quiz_data_train['Length'])
quiz_data_train_x = pd.DataFrame(quiz_data_train.drop(columns=['Length']))
quiz_data_test_y = pd.DataFrame(quiz_data_test['Length'])
quiz_data_test_x = pd.DataFrame(quiz_data_test.drop(columns=['Length']))
````

````python
# Import LinearRegression model from Scikit-Learn
from sklearn.linear_model import LinearRegression

# recalling the linear regression in a class file and making it into an object
# Create a Linear Regression object
linreg = LinearRegression()

for x_var in quiz_data_train_x:
    print('Linear Regression plot for {} againt length: \t'.format(x_var))
    x_train = pd.DataFrame(quiz_data_train_x[x_var])
    # Train the Linear Regression model
    # linreg.fit(x_values,y_values)
    linreg.fit(x_train, quiz_data_train_y)
    # Coefficients of the Linear Regression line
    print('Intercept \t: b = ', linreg.intercept_)
    print('Coefficients \t: a = ', linreg.coef_)
    #Finding plotting up the graphs

    # Formula for the Regression line
    # Prediction in Y = Mx + c
    regline_x = x_train
    regline_y = linreg.intercept_ + linreg.coef_ * x_train
    print('The regression line is: y = {}x + {}'.format(linreg.coef_,linreg.intercept_))

    # Plot the Linear Regression line
    f, axes = plt.subplots(1, 1, figsize=(16, 8))
    plt.scatter(x_train, quiz_data_train_y)
    plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
    plt.show()
````

**b) Predict Length using each of the three models on both Train and Test datasets. Plot the actual values of Length against the predicted values of Length in each case. Print the Mean Squared Errors and R^2 values in each case.**

````python
# Import LinearRegression model from Scikit-Learn
from sklearn.linear_model import LinearRegression

# recalling the linear regression in a class file and making it into an object
# Create a Linear Regression object
linreg = LinearRegression()

for x_var in quiz_data_train_x:
    print('Linear Regression plot for {} againt length: \t'.format(x_var))
    x_train = pd.DataFrame(quiz_data_train_x[x_var])
    x_test = pd.DataFrame(quiz_data_test_x[x_var])
    # Train the Linear Regression model
    # linreg.fit(x_values,y_values)
    linreg.fit(x_train, quiz_data_train_y)
    # Coefficients of the Linear Regression line
    print('Intercept \t: b = ', linreg.intercept_)
    print('Coefficients \t: a = ', linreg.coef_)
    #Finding plotting up the graphs

    # Formula for the Regression line
    # Prediction in Y = Mx + c
    regline_x = x_train
    regline_y = linreg.intercept_ + linreg.coef_ * x_train
    print('The regression line is: y = {}x + {}'.format(linreg.coef_,linreg.intercept_))
    y_train_pred = linreg.predict(x_train)
    y_test_pred = linreg.predict(x_test)
        
    # Calculating the R^2 value
    # Explained Variance (R^2)
    
    
    print('For Train: \n')
    print("Explained Variance for train(R^2) \t:", linreg.score(x_train, quiz_data_train_y))
    # Mean Squared Error (MSE)
    def mean_sq_err(actual, predicted):
          '''Returns the Mean Squared Error of actual and predicted values'''
          return np.mean(np.square(np.array(actual) - np.array(predicted)))

    mse = mean_sq_err(quiz_data_train_y, y_train_pred)
    print("Mean Squared Error (MSE) \t:", mse)
    print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
    
    print('For Test: \n')
    print("Explained Variance for test(R^2) \t:", linreg.score(x_test, quiz_data_test_y))
    mse = mean_sq_err(quiz_data_test_y, y_test_pred)
    print("Mean Squared Error (MSE) \t:", mse)
    print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))

    # Plot the Prediction against the Actual
    f, axes = plt.subplots(1, 1, figsize=(16, 8))
    plt.scatter(x_test, quiz_data_test_y)
    plt.scatter(x_test, y_test_pred, color = "r")
    plt.show()
````

**c) Comment : Which of the three uni-variate linear models is the best in terms of predicting Length? Briefly justify.**


```Diameter``` is the best to predict ```Length``` as the meansquare error ~ ```0.002``` is the smallest as compared to the other and it has the highest R^2 value

### Problem 3 : Multi-Variate Linear Regression

a) Partition the dataset randomly into Train and Test datasets  70% for Train and 30% for Test. On the Train set, fit a single multi-variate Linear Regression model for Length against all of the three other variables. 

````python
## Using Random Splitting
from sklearn.model_selection import train_test_split
quiz_data_train, quiz_data_test = train_test_split(quiz_data, test_size = 0.3)
quiz_data_train_y = pd.DataFrame(quiz_data_train['Length'])
quiz_data_train_x = pd.DataFrame(quiz_data_train.drop(columns=['Length']))
quiz_data_test_y = pd.DataFrame(quiz_data_test['Length'])
quiz_data_test_x = pd.DataFrame(quiz_data_test.drop(columns=['Length']))
````

````python
# Import LinearRegression model from Scikit-Learn
from sklearn.linear_model import LinearRegression

# recalling the linear regression in a class file and making it into an object
# Create a Linear Regression object
linreg = LinearRegression()
x_train = pd.DataFrame(quiz_data_train_x)
x_test = pd.DataFrame(quiz_data_test_x)
# Train the Linear Regression model
# linreg.fit(x_values,y_values)
linreg.fit(x_train, quiz_data_train_y)
# Coefficients of the Linear Regression line
print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
#Finding plotting up the graphs

# Formula for the Regression line
# zip() is to put the itertator together with its pair, pairing the column && value
# Making it into a list so we can convert it into a data fram

# Print the Coefficients against Predictors
pd.DataFrame(list(zip(x_train.columns, linreg.coef_[0])), columns = ["Predictors", "Coefficients"])
# Prediction in Y = Mx + c

# Predict the Total values from Predictors
y_train_pred = linreg.predict(x_train)
y_test_pred = linreg.predict(x_test)


# Plot the Predictions vs the True values
f, axes = plt.subplots(1, 2, figsize=(24, 12))
axes[0].scatter(quiz_data_train_y, y_train_pred, color = "blue")
axes[0].plot(quiz_data_train_y, quiz_data_train_y, 'r-', linewidth = 2)
axes[0].set_xlabel("True values of the Response Variable (Train)")
axes[0].set_ylabel("Predicted values of the Response Variable (Train)")
axes[1].scatter(quiz_data_test_y, y_test_pred, color = "green")
axes[1].plot(quiz_data_test_y, quiz_data_test_y, 'r-', linewidth = 2)
axes[1].set_xlabel("True values of the Response Variable (Test)")
axes[1].set_ylabel("Predicted values of the Response Variable (Test)")
plt.show()

plt.show()
````

**b) Predict Length using this model on both Train and Test datasets. Print the Mean Squared Error and R^2 values.**

````python
# Import mean_squared_error from sklearn
from sklearn.metrics import mean_squared_error

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_train, quiz_data_train_y))
print("Mean Squared Error (MSE) \t:", mean_squared_error(quiz_data_train_y, y_train_pred))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_test, quiz_data_test_y))
print("Mean Squared Error (MSE) \t:", mean_squared_error(quiz_data_test_y, y_test_pred))
print()
````
